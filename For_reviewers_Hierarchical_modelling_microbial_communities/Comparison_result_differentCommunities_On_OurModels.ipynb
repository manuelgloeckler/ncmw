# Community model

```python
from ncmw.utils import get_models
from ncmw.community import BagOfReactionsModel, ShuttleCommunityModel
from ncmw.setup_models import set_default_configs_and_snm3_medium
models = get_models("models")
```

# Bag of reaction model : This very simple approach models the community as union of all reactions and metabolites of it’s members. 

```python
bag_model = BagOfReactionsModel(models)
```

```python
_, growths, _ = bag_model.optimize()
for m, g in zip(models, growths):
    print(f"{m.id.split(sep='_')[0]} achives {g} biomass rate")
```

# Change the weights

```python
bag_model.weights = [1,10]
```

```python
_, growths, _ = bag_model.optimize()
for m, g in zip(models, growths):
    print(f"{m.id.split(sep='_')[0]} achives {g} biomass rate")
```

# Shuttle Community Model
# Here any community member is separated into their own compartments, but they can interact over an external environment through so-called “shuttle reactions.”

```python
model = ShuttleCommunityModel(models)
```

```python
_, growths, _ = model.optimize()
for m, g in zip(models, growths):
    print(f"{m.id.split(sep='_')[0]} achives {g} biomass rate")
```

# Change the weights

```python
model.weights = [1,10]
```

```python
_, growths, _ = model.optimize()
for m, g in zip(models, growths):
    print(f"{m.id.split(sep='_')[0]} achives {g} biomass rate")
```
